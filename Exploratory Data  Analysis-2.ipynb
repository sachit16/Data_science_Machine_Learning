## Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?

```python
To identify the peak travel season from the given Flight Price dataset, you can analyze the 'Date_of_Journey' feature, which represents the date of the flight journey. Here are the steps you can take and features you can analyze:

1)Extract Month and Day of the Week:
Extract the month and day of the week from the 'Date_of_Journey' feature. This will allow you to identify trends based on the month and day.

2)Analyze Price Variation:
Calculate the average flight prices for each month and day of the week. This will help you identify patterns in price variation throughout the year.

3)Visualize Trends:
Use visualizations such as line plots or bar charts to showcase the average prices over different months and days of the week. This will make it easier for your boss to understand the trends.

4)Highlight Peak Season:
Identify the months or days with the highest average prices as the peak travel season. This could be due to holidays, special events, or increased demand during certain periods.

5)Consider Additional Features:
Depending on the dataset, you may also want to consider other features such as 'Source,' 'Destination,' 'Total_Stops,' and 'Duration' to provide additional context for the peak travel season.

6)Present Findings:
Prepare a report or a presentation for your boss that includes clear visualizations and insights. Highlight the identified peak travel season and provide any relevant explanations for the trends observed. Make sure to present the information in a format that is easy to understand.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'flight_data' is your DataFrame containing the flight price data
# Convert 'Date_of_Journey' to datetime format
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

# Extract month and day of the week
df['Month'] = df['Date_of_Journey'].dt.month
df['Day_of_Week'] = df['Date_of_Journey'].dt.day_name()

# Calculate average prices by month and day of the week
average_prices = df.groupby(['Month', 'Day_of_Week'])['Price'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
for month in average_prices['Month'].unique():
    month_data = average_prices[average_prices['Month'] == month]
    plt.plot(month_data['Day_of_Week'], month_data['Price'], label=f'Month {month}')

plt.xlabel('Day of the Week')
plt.ylabel('Average Price')
plt.title('Average Flight Prices by Day of the Week')
plt.legend()
plt.show()
```

## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df['Price'] = pd.to_numeric(df['Price'].replace('[\$,]', '', regex=True), errors='coerce')

# Group by App Type and calculate the average price
average_prices = df.groupby('Type')['Price'].mean()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
# Assuming df is your DataFrame with the app data

# Remove rows with index 10472
df = df.drop(df.index[10472])

# Create a copy of the DataFrame
dff = df.copy()

# Define characters to remove
char_to_remove = ['+', ',', '$']

# Define columns to clean
cols_to_clean = ['Installs', 'Price']

# Fill NaN values with a default value before replacing characters
for cols in cols_to_clean:
    dff[cols] = dff[cols].fillna('0').astype(str)

# Remove specified characters from 'Installs' and 'Price' columns
for item in char_to_remove:
    for cols in cols_to_clean:
        dff[cols] = dff[cols].str.replace(item, '')

# Convert 'Installs' and 'Price' columns to numeric
dff['Installs'] = pd.to_numeric(dff['Installs'])
dff['Price'] = pd.to_numeric(dff['Price'], errors='coerce')

# Create a frequency table for the top 10 most popular apps
top_apps = dff.nlargest(10, 'Installs')[['App', 'Installs']]

print(top_apps)
```
